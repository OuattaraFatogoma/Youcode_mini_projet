### Explications 

Le fichier contient 65 lignes et 4 colonnes ("Pays", " Age ", "Genre", "Dépenses des clients").  

Toutes les colonnes sont de type object.  

Seul la colonne "Pays" ne contient pas de case nulle les autres contiennent des cases nulle 
(3 cases nulles pour " Age ", 1 case nulle pour "Genre" et 2 cases nulles pour "Dépenses des clients")

```python
#Calcule des medianes et moyenne des colonnes " Age " et "Dépenses des clients"

##premièrement on vas convertir les colonnes " Age " et "Dépenses des clients" en nombre (numeric)
df[" Age "] = pd.to_numeric(df[" Age "], errors="coerce")
df["Dépenses des clients"] = pd.to_numeric(df["Dépenses des clients"], errors="coerce")

age_median = df[" Age "].median()
age_moyenne = df[" Age "].mean()
print("La mediane des ages est : ", age_median)
print("La moyennes des ages est : ", age_moyenne)

depenses_clients_median = df["Dépenses des clients"].median()
depenses_clients_moyenne = df["Dépenses des clients"].mean()
print("La mediane des Dépenses des clients est : ", depenses_clients_median)
print("La moyennes des Dépenses des clients est : ", depenses_clients_moyenne)
```

## Partie 2

```python
#Lecture des données
donnes = pd.read_excel("./data-sell4all.xlsx")

#Commencons par filtrer les données et donner a chaque column le type de donnée ideal
donnes = donnes.query("Genre=='Male' or Genre=='Female'")
donnes["Dépenses des clients"] = pd.to_numeric(donnes["Dépenses des clients"], errors="coerce")
donnes[" Age "] = pd.to_numeric(donnes[" Age "], errors="coerce")
donnes["Pays"] = donnes["Pays"].astype("string")
donnes["Genre"] = donnes["Genre"].astype("string")

donnes.info()
```

```python
#Affichage des dépenses des clients par pays
##On va tout d'abord grouper par pays puis calculer la somme des dépenses des clients pour chaque pays
donnes_par_pays = donnes.groupby("Pays")["Dépenses des clients"].sum().reset_index()

#Visualisation des données des dépenses des clients par pays
plt.figure(figsize=(12,6))
plt.bar(data=donnes_par_pays, x=donnes_par_pays["Pays"], label=donnes_par_pays["Pays"], height="Dépenses des clients")
plt.xlabel('Pays')
plt.ylabel('Dépenses client (£)')
plt.title("Dépenses des clients par pays")
plt.legend()
plt.show()
```
